```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, 'r--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
```

```python
# multiplots
plt.subplot(1,2,1) #1row 2 col 1=pic1
plt.plot(x, y, 'r--')

plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

```python
# multiplots Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
# insert plotting
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # mail axis
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # insert axis

# Axes1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Axes 1')

# Axes2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X label_axes2')
axes2.set_ylabel('Y label_axes2')
axes2.set_title('Axes 2')
```

```python
# subplots
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
x = np.linspace(0, 5, 11)
y = x**2

fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, 'b')
axes[0].set_title('First Plot')

axes[1].plot(y,x, 'r')
axes[1].set_title('Second Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)

for ax in axes:
    ax.plot(x,y,'g')
    ax.set_xlabel('X - axes')
    ax.set_ylabel('Y - axes')
    ax.set_title('Title')
```

```python
# Figure size, as sprect ratio and DPI
# Figure : กว้าง
# DPI : dots-per-inch

fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(10,4))

axes.plot(x,y,'r')
axes.set_xlabel('X - axes')
axes.set_ylabel('Y - axes')
axes.set_title('Title')
```
